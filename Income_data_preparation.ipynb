```python
# add year column


# Define the function to add a 'Year' column
def add_year_column(df, date_column_name):
    # Convert the specified date column to datetime format
    df[date_column_name] = pd.to_datetime(df[date_column_name])
    # Create a new 'Year' column
    df['Year'] = df[date_column_name].dt.year
    return df
```
