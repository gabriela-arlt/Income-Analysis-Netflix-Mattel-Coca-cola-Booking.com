```python
import pandas as pd

# Using double backslashes or a raw string for the file path
barbie_income = pd.read_csv(r'C:\Users\gabb_\Gabriela Arlt CAB Projects\Project_API\income_statement_barbie.csv', encoding='UTF-8', sep=',')

# Display the DataFrame to verify it's loaded correctly
print(barbie_income.head())
```

## 1. Line Chart of Total Revenue and Net Income Over Time

```python
import matplotlib.pyplot as plt

# Ensure 'Fiscal Date Ending' is in datetime format
barbie_income['Fiscal Date Ending'] = pd.to_datetime(barbie_income['Fiscal Date Ending'])
barbie_income.sort_values('Fiscal Date Ending', inplace=True)

# Plot Total Revenue and Net Income
plt.figure(figsize=(12, 6))
plt.plot(barbie_income['Fiscal Date Ending'], barbie_income['Total Revenue'], label='Total Revenue', marker='o')
plt.plot(barbie_income['Fiscal Date Ending'], barbie_income['Net Income'], label='Net Income', marker='o')
plt.xlabel('Fiscal Date Ending')
plt.ylabel('Amount (in millions)')
plt.title('Total Revenue and Net Income Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

## 2. Bar Chart of Gross Profit, Operating Income, and Net Income

```python
# Set the fiscal date as the index for plotting
barbie_income.set_index('Fiscal Date Ending', inplace=True)

# Plot the selected columns
barbie_income[['Gross Profit', 'Operating Income', 'Net Income']].plot(kind='bar', figsize=(14, 7))
plt.xlabel('Fiscal Date Ending')
plt.ylabel('Amount (in millions)')
plt.title('Profitability Analysis: Gross Profit, Operating Income, and Net Income')
plt.legend()
plt.grid(True)
plt.show()
```

## 3. Stacked Bar Chart of Cost of Revenue, SG&A, R&D, and Operating Expenses

```python
# Plot the stacked bar chart
barbie_income[['Cost of Revenue', 'Selling, General and Administrative', 'Research and Development', 'Operating Expenses']].plot(kind='bar', stacked=True, figsize=(14, 7))
plt.xlabel('Fiscal Date Ending')
plt.ylabel('Amount (in millions)')
plt.title('Cost Structure Analysis')
plt.legend()
plt.grid(True)
plt.show()
```

## Line Chart of EBIT and EBITDA Over Time

```python
# Plot EBIT and EBITDA
plt.figure(figsize=(12, 6))
plt.plot(barbie_income.index, barbie_income['EBIT'], label='EBIT', marker='o')
plt.plot(barbie_income.index, barbie_income['EBITDA'], label='EBITDA', marker='o')
plt.xlabel('Fiscal Date Ending')
plt.ylabel('Amount (in millions)')
plt.title('EBIT and EBITDA Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

## 5. Pie Chart of Revenue Components

```python
# Sum the revenue components for the pie chart
revenue_components = barbie_income[['Gross Profit', 'Total Revenue']].sum()

# Plot the pie chart
revenue_components.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8))
plt.title('Revenue Breakdown')
plt.ylabel('')
plt.show()
```
